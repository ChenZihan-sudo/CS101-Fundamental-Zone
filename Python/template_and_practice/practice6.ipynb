```python
import pandas as pd
import numpy as np

data = {"姓名": ["AAA", "BBB", "CCC", "DDD", "EEE"], "性别": [1, 0, 0, 0, 1]}
df = pd.DataFrame(data)
df["成绩"] = [98, 87, 32, 67, 77]
print(df)


def grades(x):
    if x >= 90:
        return "A"
    elif x >= 80:
        return "B"
    elif x >= 70:
        return "C"
    elif x >= 60:
        return "D"
    else:
        return "F"


df["等级"] = df["成绩"].map(grades)
print(df)
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn import preprocessing
from matplotlib import pyplot as plt
import seaborn as sns

# Fonts for mac platform only.
plt.rcParams["font.sans-serif"] = ["Arial Unicode MS"]
plt.rcParams["axes.unicode_minus"] = False
sns.set(font="Arial Unicode MS")

data = pd.read_csv("./data/stu_data.csv")
print(data)

print(data.head(), "\n", data.shape)

data.isnull().any()
total = data.isnull().sum().sort_values(ascending=False)
print("total:\n", total)

percent = (data.isnull().sum() / data.isnull().count()).sort_values(ascending=False)
miss_data = pd.concat([total, percent], axis=1, keys=["Total", "Percent"])
miss_data.head(20)

import missingno

missingno.matrix(data)
data = data.dropna(thresh=data.shape[0] * 0.5, axis=1)
data.dropna(axis=0, how="all")
print(data)

print(data.duplicated().sum(), "\n", data.drop_duplicates())
```

```python
data.columns
id_col = ["姓名"]
cat_col = ["学历", "学校"]
cont_col = ["成绩", "能力"]
print(data[cat_col])
print(data[cont_col])

for i in cat_col:
    print(pd.Series(data[i]).value_counts())
    plt.plot(pd.Series(data[i]).value_counts())

dummies = pd.get_dummies(data[cat_col])
print("dummies:\n", dummies)
```

```python
data[cont_col].describe()

skewed_feats = data[cont_col].apply(lambda x: (x.dropna()).skew())
skewed_feats = skewed_feats[skewed_feats > 0.75]
skewed_feats = skewed_feats.index
data[skewed_feats] = np.log1p(data[skewed_feats])
print(skewed_feats)

scaled = preprocessing.scale(data[cont_col])
scaled = pd.DataFrame(scaled, columns=cont_col)
print(scaled)

m = dummies.join(scaled)
data_cleaned = data[id_col].join(m)
data_cleaned = m
print("normalized data:\n", data_cleaned)

print("变量之间的相关性：\n", data_cleaned.corr())

def corr_heat(df):
    dfData = abs(df.corr())
    plt.subplots(figsize=(9, 9))
    sns.heatmap(dfData, annot=True, vmax=1, square=True, cmap="Blues")
    plt.show()

corr_heat(data_cleaned)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

texts = ["orange banana apple grape", "banana apple apple apple", "orange grape grape"]
cv = CountVectorizer()
cv_fit = cv.fit_transform(texts)
print(cv.vocabulary_)
print(cv_fit.shape)
print(cv_fit)
print(cv_fit.toarray()  )
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

texts = [
    "orange banana apple grape",
    "banana apple apple apple",
    "orange grape grape",
    "orange apple",
]
cv = TfidfVectorizer()
cv_fit = cv.fit_transform(texts)
print(cv.vocabulary_)
print(cv_fit)
print(cv_fit.toarray())
```
