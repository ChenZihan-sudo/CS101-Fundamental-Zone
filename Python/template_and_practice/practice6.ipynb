```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()
print(digits.keys())
# 数据
print(digits.data.shape)
# 标签
print(digits.target.shape)
print(digits.images.shape)

plt.figure(figsize=(8, 1))
for i in range(10):
    plt.subplot(1, 10, i + 1)
    plt.imshow(digits.images[i], cmap="binary")
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(digits.target[i])
```

```python
import pandas as pd
import numpy as np

data = {"姓名": ["AAA", "BBB", "CCC", "DDD", "EEE"], "性别": [1, 0, 0, 0, 1]}
df = pd.DataFrame(data)
df["成绩"] = [98, 87, 32, 67, 77]
print(df)


def grades(x):
    if x >= 90:
        return "A"
    elif x >= 80:
        return "B"
    elif x >= 70:
        return "C"
    elif x >= 60:
        return "D"
    else:
        return "F"


df["等级"] = df["成绩"].map(grades)
print(df)
```

```python
import warnings
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
import missingno as msno
import matplotlib.pyplot as plt
from itertools import product

warnings.filterwarnings("ignore")


def getData():
    x, y = gen.make_classification(
        n_samples=1000,
        n_features=10,
        n_classes=2,
        n_clusters_per_class=1,
        random_state=0,
    )
    for i, j in product(range(x.shape[0]), range(x.shape[1])):
        if np.random.random() >= 0.8:
            xloc = np.random.randint(0, 10)
            x[i, xloc] = np.nan
    return x, y


plt.scatter(x[:, 0], x[:, 1], marker="o", c=y)

x, y = getData()
df = pd.DataFrame(x, columns=["x%s" % str(i) for i in range(x.shape[1])])
df["label"] = y
msno.matrix(df)
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "cat fish cat cat cat cat cat dog dog dog fish duck duck"
wordcloud = WordCloud(
    font_path="Arial.ttf",
    max_font_size=500,
    width=1000,
    height=400,
    scale=2,
    random_state=50,
)
wordcloud.generate(text)

plt.imshow(wordcloud, interpolation="none")
plt.axis("off")
```
